```python
import numpy as np
# prompt: fixed point iteration code
# this function finds a fixed point of the function
# g(x) using fixed-point iteration
# INPUTS:
# g: The function for which to find a fixed point (g(x) = x).
# x0: The initial guess for the fixed point.
# tolerance: The desired accuracy of the solution.
# max_iterations: The maximum number of iterations to perform.
import math
def fixed_point_iteration(g, x0, tolerance=1e-3, max_iterations=100):
  x_prev = x0
  print("The initial guess is {}".format(x_prev))
  for i in range(max_iterations):
    x_new = g(x_prev)
    if abs(x_new - x_prev) < tolerance:
      return x_new
    x_prev = x_new
    print("The {}th iteration is {}".format(i+1, x_new))
  return None
```

Function 1: $g(x) = 1-.5(x^3)$



```python
g = lambda x: 0.7704615886883557-.5*x**3
fixed_point_iteration(g,0.7704615886883557)
```

$x_0 = 1$

$x_1 = x_0 -.5(x_1)^3 \implies x_1 = .77$

$x_2 = x_1-.5(x_2)^3 = .77 -.5(x_2)^3 \implies x_2 = .64$

Function 2: $g_2(x) = [(10/x)-4x]^{1/2}$

```python
g2 = lambda x: np.sqrt(((10/x)-4*x))
fixed_point_iteration(g2, 1.5)
```

Function 3: $g_3(x) = 1/2(10-x^3)^{1/2}$

```python
g3 = lambda x: 1/2*np.sqrt((10-x**3))
fixed_point_iteration(g3, 1.5)
```

Function 4: $g_4(x) = (10/(4+x))^{1/2}$

```python
g4 = lambda x: np.sqrt((10/(4+x)))
fixed_point_iteration(g4, 1.5)
```

Function 5: $g_5(x) = x-\frac{x^3+4x^2-10}{3x^2+8x}$
