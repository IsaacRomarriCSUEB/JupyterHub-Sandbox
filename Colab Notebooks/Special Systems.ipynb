$A = R^TR$ where R is upper triangular

$A = LL^T$ where L is lower triangular

```python
import numpy as np
A = np.array([[1,2,6],[2,5,15],[6,15,46]])
L = np.linalg.cholesky(A)
print("L = ",L)
```

```python
print(np.dot(L,L.T))
```
